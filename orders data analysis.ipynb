```python
#extract the zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() #extract file to folder
zip_ref.close() #close connection
```

```python
import pandas as pd
df = pd.read_csv('orders.csv', na_values = ['Not Available', 'unknown']) #Making Not Available,unknown values as null values(nan) 
df.head(20)
df['Ship Mode'].unique() #to get all the unique values in Ship Values
```

```python
#renaming coulmn names....make them lower case and replacing space with underscore
#df.rename(columns={'Order Id':'order_id','City':'city'}) manual replacing
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
df.head()
```

```python
#derive new columns discount, sale_price and profit
df['discount_price'] = df['list_price']*df['discount_percent']*.01
df['sale_price'] = df['list_price'] - df['discount_price']
df['profit'] = df['sale_price'] - df['cost_price']
df.head(20)
```

```python
#convert order date from object datatype to datetime
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df.dtypes
```

```python
#drop cost_price, list_price and discount_price columns
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
df.head(10)
```

```python
#load the data into sql server using replace option
import sqlalchemy as sal

engine = sal.create_engine(r'mssql://Lighting_Speed\SQLEXPRESS/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn = engine.connect()
```

```python
df.to_sql('df_orders', con=conn , index=False, if_exists = 'append')
```
